```python
def remove_outlier(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range 
```

```python
for i in num:    
    LL,UL=remove_outlier(customer[i])
    print("Proportion of outlier in ", i ,"is",round(len(customer[(customer[i]>UL)|(customer[i]<LL)])/len(customer[i]),3))
    print()
```

```python
for column in num:
    lr,ur=remove_outlier(customer[column])
    customer[column]=np.where(customer[column]>ur,ur,customer[column])
    customer[column]=np.where(customer[column]<lr,lr,customer[column])
```

### Finding optimum threshold 

```python
# calculate AUC
auc = metrics.roc_auc_score(y_train,pred_prob_train[:,1])
print('AUC for the Training Data: %.3f' % auc)

#  calculate roc curve
fpr, tpr, thresholds = metrics.roc_curve(y_train,pred_prob_train[:,1])
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.',label = 'Training Data')
```

```python
# roc curve for logistic regression model with optimal threshold
from numpy import argmax
```

```python
J = tpr - fpr
ix = argmax(J)
best_thresh = thresholds[ix]
print('Best Threshold=%f' % (best_thresh))
```

### Fitting the optimum threshold on the datasetÂ¶

```python
y_class_pred_train=[]
for i in range(0,len(pred_prob_train)):
    if np.array((pred_prob_train)[:,1])[i]>best_thresh:
        a=1
    else:
        a=0
    y_class_pred_train.append(a)
```

```python
sns.heatmap((metrics.confusion_matrix(y_train,y_class_pred_train)),annot=True,fmt='.5g'
            ,cmap='Blues');
plt.xlabel('Predicted');
plt.ylabel('Actuals',rotation=0);
```

```python
print(metrics.classification_report(y_train,y_class_pred_train,digits=3))
```

```python
y_class_pred_test=[]
for i in range(0,len(pred_prob_test)):
    if np.array((pred_prob_test)[:,1])[i]>best_thresh:
        a=1
    else:
        a=0
    y_class_pred_test.append(a)
```

```python
sns.heatmap((metrics.confusion_matrix(y_test,y_class_pred_test)),annot=True,fmt='.5g'
            ,cmap='Blues');
plt.xlabel('Predicted');
plt.ylabel('Actuals',rotation=0);
```
